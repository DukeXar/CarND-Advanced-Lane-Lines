```python
from camera import *

def test_calibration():
    import cv2
    from matplotlib import pyplot as plt

    grid_shape = (9, 6)
    images = load_calibration_images('./camera_cal')
    mtx, dist, rvecs, tvecs, imgpoints, images = calibrate_camera(images, grid_shape)
    for img, corners in zip(images, imgpoints):
        img_with_corners = img.copy()
        cv2.drawChessboardCorners(img_with_corners, grid_shape, corners, True)

        undistorted = cv2.undistort(img, mtx, dist, None, mtx)

        fig, axes = plt.subplots(1, 2, figsize=(12, 6))

        # Use [:,:,::-1] to flip BGR to RGB
        axes[0].set_title('Original')
        axes[0].imshow(img_with_corners[:,:,::-1], interpolation='bicubic')
        axes[1].set_title('Undistorted')
        axes[1].imshow(undistorted[:,:,::-1], interpolation='bicubic')

        fig.tight_layout()
        plt.show()
```

```python
test_calibration()
```
